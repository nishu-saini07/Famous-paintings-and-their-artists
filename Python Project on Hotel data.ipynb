# Data analysis and visualization


```python
data
```

```python
cancel_room=data['is_canceled'].value_counts()
print(cancel_room)
```

```python
plt.figure(figsize=(5,4))
plt.title("Booking status count",size=15)
plt.bar(['not cancel','cancel'],data['is_canceled'].value_counts(),color='Orange')
plt.show()
```

```python
plt.figure(figsize=(10,4))
#is_canceled=data['is_canceled'].value_counts()
plt.title("Hotel count based on cancellation",size=15)
ax=sns.countplot(x='hotel',hue='is_canceled',data=data)

plt.xlabel('Hotels')
plt.ylabel('Number of Booking')
plt.legend(['Not canceled','canceled'])
plt.show()
        
```

```python
resort_hotel=data[data['hotel']=='Resort Hotel']
x=resort_hotel['is_canceled'].value_counts(normalize = True)
fig=plt.figure(figsize=(5,5))
axes=fig.add_axes([0.2,0.2,0.9,1])
plt.pie(x,autopct=' %.2f')
plt.title("Booking % status of Resort Hotel",size=15)
plt.legend([' Not canceled',' canceled'])
plt.show()
```

```python
city_hotel=data[data['hotel']=='City Hotel']
y=city_hotel['is_canceled'].value_counts(normalize = True)
fig=plt.figure(figsize=(5,5))
axes=fig.add_axes([0.2,0.2,0.9,1])
plt.pie(y,autopct=' %.2f')
plt.title("Booking % status of City Hotel",size=15)
plt.legend([' Not canceled',' canceled'])
plt.show()
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Average daily price in City and Resort Hotel",size=15)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend()
plt.show()
```

```python
canceled_hotel=data[data['is_canceled']==1]
not_canceled_hotel=data[data['is_canceled']==0]
canceled=canceled_hotel.groupby(['reservation_status_date'])[['adr']].mean()
canceled.reset_index(inplace=True)
canceled.sort_values('reservation_status_date',inplace=True)
not_canceled=not_canceled_hotel.groupby(['reservation_status_date'])[['adr']].mean()
not_canceled.reset_index(inplace=True)
not_canceled.sort_values('reservation_status_date',inplace=True)
plt.figure(figsize=(20,8))
plt.plot(canceled['reservation_status_date'],canceled['adr'],label='canceled')
plt.plot(not_canceled['reservation_status_date'],not_canceled['adr'],label='Not canceled')
plt.title('Average daily Rate',size=20)
plt.legend(fontsize=15)
plt.show()
```

```python
canceled=canceled[(canceled['reservation_status_date']>'2016')&(canceled['reservation_status_date']<'2017-09')]
not_canceled=not_canceled[(not_canceled['reservation_status_date']>'2016')&(not_canceled['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate',fontsize=20)
plt.plot(canceled['reservation_status_date'],canceled['adr'],label='canceled')
plt.plot(not_canceled['reservation_status_date'],not_canceled['adr'],label='Not canceled')
plt.legend(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
data['month']=data['reservation_status_date'].dt.month
plt.bar('month','adr',data=data[data['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(),color='orange')
plt.title('Price per month ',size=15)
plt.xlabel('Month')
plt.ylabel('Price')
plt.show()
```

```python
data['month']=data['reservation_status_date'].dt.month
plt.figure(figsize=(10,5))
sns.countplot(x='month',hue='is_canceled',data=data)
plt.title('Booking status per month')
plt.xlabel('Month')
plt.ylabel('Number of booking')
plt.legend(['Not canceled','Canceled'])
plt.show()
```

```python
plt.figure(figsize=(9,4))
data['year']=data['reservation_status_date'].dt.year
plt.plot('year','is_canceled',data=data[(data['hotel']=='Resort Hotel') & (data['is_canceled']==1)].groupby('year')
         [['is_canceled']].count().reset_index(),marker='o')
plt.plot('year','is_canceled',data=data[(data['hotel']=='City Hotel') & (data['is_canceled']==1)].groupby('year')
         [['is_canceled']].count().reset_index(),marker='o')
plt.legend(['Resort Hotel','City Hotel'])
plt.title('Total Cancellation by year',size=15)
plt.xlabel('Year')
plt.ylabel('No of cancellation')
plt.show()
```

```python
canceled=data[data['is_canceled']==1]
top_10_countries=canceled['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.pie(top_10_countries,autopct='%.2f',labels=top_10_countries.index)
plt.title('Reservation cancellation in Top 10 countries',size=15)
plt.show()
```
